```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating synthetic daily temperature data
np.random.seed(0)
days = pd.date_range(start="2024-10-01", end="2024-10-31", freq="D")
temps = np.random.normal(loc=25, scale=5, size=len(days))
```

# Dataset for predicting ongoing month temperature

```python
# Creating a DataFrame
df = pd.DataFrame({'Date': days, 'Temperature': temps})
df
```

```python
# Analyzing the data
print(f"Mean Temperature: {df['Temperature'].mean():.2f}°C")
print(f"Max Temperature: {df['Temperature'].max():.2f}°C")
print(f"Min Temperature: {df['Temperature'].min():.2f}°C")
```

```python
# Visualizing the data
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Temperature', data=df)
plt.title('Daily Temperatures in October 2024')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Temperature'], bins=10, kde=True)
plt.title('Distribution of Daily Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

# This example illustrates how data science can take raw data, analyze it for insights. & visualize it to understand patterns better.
